```python
def convert_to_numeric(x):
    try:
        mean_price = df[df[x]!='?'][x].astype(int).mean()
        df.loc[df[x]== '?',x] = mean_price
    except:
        pass
    df[x] = df[x].astype(int)
    
```

```python
df.loc[df['price']== '?','price']  = df[df['price']!='?']['price'].astype(int).mean()
```
